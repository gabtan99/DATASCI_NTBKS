```
import requests

URL="https://edition.cnn.com/article/sitemap-2021-3.html"
```

```
page=requests.get(URL)
```

```
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'html.parser')

all_news = soup.find_all("span", {"class": "sitemap-link"})

all_links = []
for html in all_news:
  soup = BeautifulSoup("'" + str(html) + "'", 'html.parser')
  a = soup.find("a")
  
  if a is not None:
    if "2021/03/11" in str(a["href"]) or "2021/03/12" in str(a["href"]):
      all_links.append(a["href"])
    
print(all_links)
```

```
import re
import json
all_json = []

for link in all_links:
  page = requests.get(link)
  soup = BeautifulSoup(page.content, 'html.parser')

  title = soup.find("h1", {"class": "pg-headline"}).text.strip()

  date_full = soup.find("p", {"class": "update-time"}).text.replace('Updated', '').strip()
  date = re.sub(r"\([^()]*\)", "", date_full)

  author = soup.find("span", {"class": "metadata__byline__author"}).text.replace('By', '').strip()

  start = soup.find("p", {"class": "zn-body__paragraph speakable"}).text.replace('(CNN)', '').strip()
  full_article =  start

  for para in soup.find_all("div", {"class": "zn-body__paragraph"}):
    full_article = full_article + " " + para.text.strip()

  all_json.append({
      "date": date,
      "title": title,
      "full_article": full_article,
      "author": author
  })

print(all_json)

with open('news.json', 'w') as outfile:
    json.dump(all_json, outfile)
```

```
# use this to pretty print the json to view its format

with open('news.json', 'r') as handle:
    parsed = json.load(handle)
    print(json.dumps(parsed, indent=4, sort_keys=True))
```
